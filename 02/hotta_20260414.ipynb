- dir()
    - 引数がない場合、現在のローカルスコープにある名前のリストを返します。

- type()

## 演習2_2

````python
x = 3
print(f"type(x): {type(x)}")
print(f"id(x): {id(x)}")
print(f"dir(x): {dir(x)}")
````

````python
y = x
print(f"比較")
print(f"id(x): {id(x)}")
print(f"id(y): {id(y)}")
````

````python
y = y - 1
print(f"比較")
print(f"x: {x}")
print(f"id(x): {id(x)}")
print(f"y: {y}")
print(f"id(y): {id(y)}")
````

変数yの値を変更するとidも変更される。一方xの値とid(x)は変更されなかった。yは新しいオブジェクトを参照するようになったと言える

## 演習2_7

DataFrame から列を抽出する際、`pearson1["Father"]` と `pearson1[["Father"]]` には以下のような違いがある。

### 1. 返ってくるオブジェクトの型が異なる

- `pearson1["Father"]` → **Series**（1次元のデータ構造）として返される
- `pearson1[["Father"]]` → **DataFrame**（2次元のデータ構造）のまま返される

```python
type(pearson1["Father"])    # pandas.core.series.Series
type(pearson1[["Father"]])  # pandas.core.frame.DataFrame
```

### 2. 利用できる属性・メソッドが異なる

- Series には `.name`, `.values`, `.index` などがある一方、`.columns` や 2次元的な `.shape` は持たない
- DataFrame は `.columns`, `.shape`（行数×列数の2次元）などを持ち、複数列を扱う前提のメソッドが使える

### 3. 表示形式が異なる

- Series は列名が上部に1つだけ表示され、縦にデータが並ぶ
- DataFrame は表形式（列名が横並び）で表示される

### 4. 複数列を抽出したい場合は二重かっこが必須

- `pearson1[["Father", "Son"]]` のように書くと複数列の DataFrame を取得できる
- `pearson1["Father", "Son"]` はエラーになる（タプルキーと解釈されるため）
