```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
dir(iris)
```

```python
print(iris.DESCR)
```

```python
print(iris.data)
```

```python
print(iris.feature_names)
```

```python
print(iris.filename)
```

```python
print(iris.target)
```

```python
print(iris.target_names)
```

```python
import pandas as pd
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df
```

```python
iris_df[10:21]
```

```python
iris_df[-10:]
```

```python
iris_df.shape
```

```python
iris_df.describe()
```

```python
iris_df.mean()
```

```python
iris_df["sepal length (cm)"].mean()
```

```python
iris_df[["sepal length (cm)", "sepal width (cm)"]].mean()
```

```python
# 標準偏差の確認
print('標準偏差')
print(iris_df.std())
print() # 1行の空行を挿入するため

# 最小値の確認
print('最小値')
print(iris_df.min())
print()

#　最大値の確認
print('最大値')
print(iris_df.max())
print()

#　中央値の確認
print('中央値')
print(iris_df.median())
print()

#　データ数の確認
print('データ数')
print(iris_df.count())
print()

#　25%点の確認
print('25%点')
print(iris_df.quantile(0.25))
print()
```
