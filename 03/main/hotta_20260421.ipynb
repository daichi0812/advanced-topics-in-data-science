#### 演習課題2

演習課題2-1

```python
breast_cancer = datasets.load_breast_cancer()
pd.set_option("display.max_columns", None)
breast_df = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
breast_df.head()
```

```python
breast_df.describe()
```

演習課題2-2

```python
large_cols = [3, 23]  # mean area, worst area

breast_df.iloc[:, large_cols].boxplot()
plt.show()

other_cols = [i for i in range(breast_df.shape[1]) if i not in large_cols]
breast_df.iloc[:, other_cols].boxplot(figsize=(20, 6), rot=90)
plt.savefig("breast_boxplot_others.png", dpi=144)
plt.show()
```

### ヒストグラム

```python
iris_df.hist()    # ヒストグラムの作成
plt.show()        # ヒストグラムの表示
```

```python
iris_df.hist(bins=20)    # ビン数を20にしたヒストグラムの作成
plt.show()               # ヒストグラムの表示
```

```python
iris_df['species'] = [iris.target_names[i] for i in iris.target]
```

```python
iris_df.head()
```

```python
# column で集計する変数を指定
# by でクラス情報が記述されている列を指定
# sharex はすべてのグラフで x軸の範囲を揃えたいときに True にする
# sharey はすべてのグラフで y軸の範囲を揃えたいときに True にする
# layout は個々のグラフの配置を指定する．(3,1)は3行1列の意味
iris_df.hist(column='sepal length (cm)', by='species', sharex=True, sharey=True, layout=(3,1)) 
plt.show()               # ヒストグラムの表示
```

### 散布図

```python
iris_df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)')   # 散布図の作成
plt.show()      # 散布図の表示
```

```python
iris_df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', alpha=0.5)  # 点の透過度を半減させた散布図の作成
plt.show()       # 散布図の表示
```

```python
ax = plt.gca()           # 軸オブジェクトの取得
ax.set_aspect('equal')   # アスペクト比の指定
iris_df.plot.scatter(x='sepal length (cm)', 
                     y='sepal width (cm)',
                     alpha=0.5,
                     ax=ax)  # 点の透過度を半減させた散布図の作成
plt.show()       # 散布図の表示
```
